```python
import pandas as pd

df = pd.read_csv('../data/2019-Oct.csv')

print("Satır sayısı:", df.shape[0])
print("Sütun sayısı:", df.shape[1])
df.head(10)
```

```python
df.info()
```

```python
print(df['event_type'].value_counts())
print("\nOran (%):")
print((df['event_type'].value_counts() / len(df) * 100).round(2))
```

```python
print("Eksik değerler:")
print(df.isnull().sum())
print("\nEksik oran (%):")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
print("Benzersiz kullanıcı:", df['user_id'].nunique())
print("Benzersiz ürün:", df['product_id'].nunique())
print("Benzersiz marka:", df['brand'].nunique())
```

```python
purchases = df[df['event_type'] == 'purchase']
print("En çok satın alınan markalar:")
print(purchases['brand'].value_counts().head(10))
```

```python
import plotly.express as px

df['event_time'] = pd.to_datetime(df['event_time'])
df['date'] = df['event_time'].dt.date

daily = df[df['event_type'] == 'purchase'].groupby('date').size().reset_index()
daily.columns = ['date', 'purchase_count']

fig = px.line(daily, x='date', y='purchase_count', title='Ekim 2019 - Günlük Satın Alma Sayısı')
fig.show()
```
